## Logistic Map Widget

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
%matplotlib inline

def logistic_map(lamda):
    N = 100
    x = np.zeros(N)
    x[0] = np.random.random()
    for n in range(N-1):
        x[n+1] = lamda*x[n]*(1-x[n])
    plt.figure(figsize=(10, 5))
    plt.clf()
    plt.plot(x, 'b.-')
    plt.ylim(0, 1)
    plt.xlabel('Number of iterations')
    plt.ylabel('x')
    plt.show()

# Create a FloatSlider for lambda with more precision and larger width
lamda_slider = widgets.FloatSlider(
    value=0, 
    min=3.4, 
    max=3.57, 
    step=0.0001, 
    description='Lambda:', 
    readout_format='.4f',
    style={'description_width': 'initial'}, # This ensures the description does not get cut off
    layout=widgets.Layout(width='1000px') # Adjust the width as needed
)

widgets.interact(logistic_map, lamda=lamda_slider)
```

## Logistic Map Parameter Sweep

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget

def logistic_map(r, Nt, Nx):
    x = np.random.random()
    for i in range(Nt):
        x = r * x * (1 - x)

    s = np.zeros(Nx)
    for i in range(Nx):
        x = r * x * (1 - x)
        s[i] = x
    return s

Nr = 2**10 # Number of r values
Nx = 2**10 # Number of cycle iterations
Nt = 2**14 # Number of transient iterations
x = np.zeros((Nr, Nx))
r = np.linspace(3.5, 3.9, Nr)

for i in range(Nr):
    x[i] = logistic_map(r[i], Nt, Nx)

plt.figure(figsize=(10, 5))
# plt.clf()
plt.plot(r, x, 'b.', markersize=0.1)
plt.xlabel('r')
plt.ylabel('Settled values of x')
plt.show()
```

## Widget plot of $f^{2n}(x)$

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
%matplotlib inline

# Define the logistic map function
def logistic_map(lamda):
    N = 10000
    x = np.linspace(0, 1, N)
    f = x
    plt.figure(figsize=(10, 5))
    plt.clf()
    for n in range(16):
        f = lamda * f * (1 - f)
        if np.log2(n+1) % 1 == 0:
            plt.plot(x, f, label = f'n = {n+1}')
            plt.plot(x, x)

    x0 = np.random.rand()
    for n in range(1000):
        x0 = lamda * x0 * (1 - x0)
        if n > 900:
            plt.plot(x0, x0, 'ro')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.title('Logistic map')
    plt.show()

# Create a FloatSlider for lambda with more precision and larger width
lamda_slider = widgets.FloatSlider(
    value=2.9, 
    min=2.9, 
    max=3.57, 
    step=0.0001, 
    description='Lambda:', 
    readout_format='.4f',
    style={'description_width': 'initial'}, # This ensures the description does not get cut off
    layout=widgets.Layout(width='1000px') # Adjust the width as needed
)

# Display the plot with the interactive slider
widgets.interact(logistic_map, lamda=lamda_slider)
```

## Renormalization widget

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets

def renormalization(alpha, rc):

    # alpha = -2.50290787509589282228390287321821578638 # https://oeis.org/A006891
    # rc = 3.56994567187094490184200515138649893676

    Ni = 32
    Nx = 10000
    x = np.linspace(0, 1, Nx)
    xm = 0.5

    f1 = x
    for n in range(Ni):
        f1 = rc * f1 * (1 - f1)
    x1 = x - xm
    f1 = f1 - xm

    f2 = x
    for n in range(2*Ni):
        f2 = rc * f2 * (1 - f2)
    x2 = x - xm
    f2 = f2 - xm

    plt.figure(figsize=(10, 5))
    plt.clf()
    plt.plot(x1, f1)
    plt.plot(x2*alpha, f2*alpha)
    plt.xlim(-0.1, 0.1)
    plt.ylim(-0.1, 0.1)
    plt.show()

# Create a FloatSlider for lambda with more precision and larger width
alpha_slider = widgets.FloatSlider(
    value=-2.50290787509589282228390287321821578638, 
    min=-2.6, 
    max=-2.4, 
    step=0.0001, 
    description='Lambda:', 
    readout_format='.4f',
    style={'description_width': 'initial'}, # This ensures the description does not get cut off
    layout=widgets.Layout(width='500px') # Adjust the width as needed
)

# Create a FloatSlider for lambda with more precision and larger width
rc_slider = widgets.FloatSlider(
    value=3.56994567187094490184200515138649893676, 
    min=3.56, 
    max=3.58, 
    step=0.0001, 
    description='Lambda:', 
    readout_format='.4f',
    style={'description_width': 'initial'}, # This ensures the description does not get cut off
    layout=widgets.Layout(width='500px') # Adjust the width as needed
)

# Display the plot with the interactive slider
widgets.interact(renormalization, alpha=alpha_slider, rc=rc_slider)
```

## Lyapunov Exponents

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget

def logistic_map(r, x):
    return r * x * (1 - x)

def calculate_lyapunov_exponent(r, initial_x=0.5, num_iterations=10000):
    x = initial_x
    for i in range(num_iterations):
        x = logistic_map(r, x)
        
    lyapunov_sum = 0
    for i in range(num_iterations):
        x = logistic_map(r, x)
        if r - 2 * r * x != 0:
            lyapunov_sum += np.log(abs(r - 2 * r * x))

    return lyapunov_sum / num_iterations

rc = 3.56994567187094490184200515138649893676 # https://oeis.org/A098587
Nr = 1000

# General sweep
# r = np.linspace(0.1, 3.9, Nr)

# Sweep around rc linearly
# rx = np.linspace(-1e-7, 1e-7, Nr)
# r = rc + rx

# Approach to rc exponentially
rx = np.linspace(0, 15, Nr)
r = rc - np.exp(-rx)

# Approach to rc exponentially from both sides
# rx = np.linspace(-15, 0, Nr)
# r = rc - np.exp(rx[::-1])
# r = np.concatenate((r,rc + np.exp(rx)))
# rx = np.concatenate((rx,rx+15))

lyapunov_exponents = np.zeros(len(r))
for i in range(len(r)):
    lyapunov_exponents[i] = calculate_lyapunov_exponent(r[i])

plt.figure(figsize=(10, 5))
plt.clf()
plt.plot(rx, lyapunov_exponents, 'b.-', markersize=0.1)
plt.xlabel('r')
plt.ylabel('Lyapunov exponent')
plt.show()
```

## Binary search for rc

```python
import numpy as np

def logistic_map(r, x):
    return r * x * (1 - x)

def calculate_lyapunov_exponent(r, initial_x=0.5, num_iterations=2**15):
    x = initial_x
    for i in range(num_iterations):
        x = logistic_map(r, x)
        
    lyapunov_sum = 0
    for i in range(num_iterations):
        x = logistic_map(r, x)
        lyapunov_sum += np.log(abs(r - 2 * r * x))

    return lyapunov_sum / num_iterations

def binary_search_for_chaos(r_min, r_max, precision):
    lower_bound = r_min
    upper_bound = r_max
    best_r = 0

    while upper_bound - lower_bound > precision:
        mid_r = (lower_bound + upper_bound) / 2
        lyapunov_exp = calculate_lyapunov_exponent(mid_r)

        if lyapunov_exp < 0:
            lower_bound = mid_r
        else:
            upper_bound = mid_r
            best_r = mid_r

    return best_r

# Set the parameters for the search
r_min = 3.569
r_max = 3.570
precision = 1e-14

onset_of_chaos_r = binary_search_for_chaos(r_min, r_max, precision)

rc = 3.56994567187094490184200515138649893676 # https://oeis.org/A098587

print(f"Estimated value of the onset of chaos  = {onset_of_chaos_r}")
print(f"Exact value of the onset of chaos is r = {rc}")
print(f"Difference from exact value = {rc - onset_of_chaos_r}")
```

## Binary search for Feigenbaum constants

```python
import numpy as np
from decimal import Decimal, getcontext

# Set the precision to 32 decimal places
getcontext().prec = 32
tolerance=Decimal('1e-31')

def find_bifurcation_point(a, b, n_iter, tolerance=tolerance):
    low = Decimal(a)
    high = Decimal(b)
    while high - low > tolerance:
        mid1 = low + (high - low) / Decimal('3')
        mid2 = high - (high - low) / Decimal('3')
        f_mid1 = derivative_at_fixed_point(mid1, n_iter)
        f_mid2 = derivative_at_fixed_point(mid2, n_iter)
        if f_mid1 > f_mid2:
            high = mid2
        else:
            low = mid1
    return (low + high) / 2

def logistic_iteration(x, r, n_iter):
    x = Decimal(x)
    r = Decimal(r)
    df = Decimal('1')
    for n in range(n_iter):
        df *= r * (1 - 2 * x)
        x = r * x * (1 - x)
    return x, df

def find_fixed_point(r, n_iter, tolerance=tolerance):
    low = Decimal('0')
    high = Decimal('1')
    while high - low > tolerance:
        mid = (low + high) / 2
        f_mid, _ = logistic_iteration(mid, r, n_iter)
        if f_mid < mid:
            high = mid
        else:
            low = mid
    return mid

def derivative_at_fixed_point(r, n_iter):
    xf = find_fixed_point(r=r, n_iter=n_iter)
    _, df = logistic_iteration(xf, r=r, n_iter=n_iter)
    return df

def find_min_d(r, n_iter):
    min_d = Decimal('1')
    xf = find_fixed_point(r=r, n_iter=n_iter)
    for n in range(n_iter):
        if abs(xf - Decimal('0.5')) < min_d:
            min_d = abs(xf - Decimal('0.5'))
        xf, _ = logistic_iteration(xf, r, n)
    return min_d

# Using Decimal for constants
feigenval1 = Decimal('4.669')
r1 = Decimal('3.0')
d1 = find_min_d(r1, 2**1)
r2 = find_bifurcation_point(Decimal('3.4'), Decimal('3.5'), 2**2)
d2 = find_min_d(r2, 2**2)
for n in range(3, 11):
    delta_r = (r2 - r1) / feigenval1
    r3_low = r2 + delta_r - delta_r/feigenval1
    r3_high = r2 + delta_r + delta_r/feigenval1
    r3 = find_bifurcation_point(r3_low, r3_high, 2**n)
    d3 = find_min_d(r3, 2**n)
    feigenval1 = (r2 - r1) / (r3 - r2)
    feigenval2 = (d2 - d1) / (d3 - d2)
    print(2**n, r3, feigenval1, feigenval2)
    r1 = r2
    r2 = r3
    d1 = d2
    d2 = d3
```
