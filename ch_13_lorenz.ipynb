## Duffing oscillator: Poincare map

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
%matplotlib inline

T = 1000000
dt = 0.01
w = 1
F = 0.4
b = 0.25

x = np.zeros(T)
v = np.zeros(T)
t = np.arange(0,T*dt,dt)

x[0] = 0.1
v[0] = 0.1

for n in range(T-1):
    a = F*np.cos(w*n*dt) - b*v[n] + x[n] - x[n]**3

    x[n+1] = x[n] + v[n]*dt
    v[n+1] = v[n] + a*dt

for n in range(10):
    ts = abs((t + n/100*2*np.pi) % (2*np.pi)) < dt
    clear_output(wait=True)
    plt.clf()
    plt.plot(x[ts],v[ts], 'b.')
    plt.xlim(-2,2)
    plt.ylim(-2,2)
    plt.show()
```
